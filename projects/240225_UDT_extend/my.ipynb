```python
# functions 

import re
import numpy as np
from collections import defaultdict

# parsing attention score and return lists with 6 statistics information
def attention_distribution(input_string):
    values = []
    for x in input_string.split():
        if 'tensor(' in x:
            values.append(float(re.findall("\d+\.\d+", x)[0]))
    
    y = defaultdict()
    y['original'] = values[:10]
    y['features'] = values[10:20]
    y['structure'] = values[20:]

    result = [
        np.median(y['original']), np.mean(y['original']),
        np.median(y['features']), np.mean(y['features']),
        np.median(y['structure']), np.mean(y['structure'])
    ]

    return result


data = defaultdict(list)
columns = [
    'att original (median)', 'att original (mean)', 
    'att features (median)', 'att features (mean)', 
    'att structure (median)', 'att structure (mean)'
]

indicies = ['texas', 'cornell', 'wisconsin', 'chameleon', 'squirrel', 'actor']

# append statistics information to dataframe dictionary 
def attention_dist_dataframe(x):
    y = attention_distribution(x)

    for idx in range(len(y)):
        key = columns[idx]
        value = y[idx]
        data[key].append(value)
```
